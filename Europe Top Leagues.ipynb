```python
# Distribusi Jumlah Gol per Pertandingan
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(df['home_score_full_time'] + df['away_score_full_time'], bins=10, kde=True, palette="viridis")
plt.xlabel("Number of Goals in The Match")
plt.ylabel("Frequency")
plt.title("Distribution of Number of Goals per Match")
plt.show()
```

```python
#Distribusi dari Hasil Pertandingan 

plt.figure(figsize=(8,5))
sns.countplot(x=df['winner'])
plt.xticks(rotation=90)
plt.xlabel("Winner")
plt.ylabel("Number of Matches")
plt.title("Distribution of Match Results")
plt.show()
```

```python
# Menghapus "DRAW" dari daftar tim pemenang
df_wins_filtered = df[df['winner'] != 'DRAW']['winner'].value_counts()

# Visualisasi histogram hanya dengan tim pemenang, tanpa "DRAW"
plt.figure(figsize=(12,6))
sns.barplot(x=df_wins_filtered.index, y=df_wins_filtered.values, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.title("Distribution of Team Wins (Without Draw)")
plt.show()
```

```python
# Menampilkan jumlah kemenangan per tim (tanpa "DRAW")
df_wins = df[df['winner'] != 'DRAW']['winner'].value_counts()

# Menampilkan distribusi hasil pertandingan (Home Wins, Away Wins, Draws)
match_results = df['winner'].copy()
match_results[df['winner'] == df['home_team']] = 'Home Win'
match_results[df['winner'] == df['away_team']] = 'Away Win'
match_results[df['winner'] == 'DRAW'] = 'Draw'

# Visualisasi distribusi hasil pertandingan
plt.figure(figsize=(8,5))
sns.countplot(x=match_results, order=['Home Win', 'Draw', 'Away Win'], palette="viridis")
plt.xlabel("Match Results")
plt.ylabel("Number of Matches")
plt.title("Distribution of Match Results")
plt.show()

# Menampilkan 10 tim dengan kemenangan terbanyak
df_wins.head(10)
```
